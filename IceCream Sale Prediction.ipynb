```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("icecream_sales.csv")
```

```python
df.head()
```

```python
#plot graph for visualise the data
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.scatter(df.Temperature,df.Sales,color="Blue",marker="*")
plt.show()
```

```python
new_df=df.drop('Sales',axis='columns')
new_df
```

```python
new_sale=df.Sales
new_sale
```

```python
#train the regression model
reg=LinearRegression()
```

```python
reg.fit(new_df,new_sale)
```

```python
reg.predict([[33]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#here we verfiy our prediction by formula Y=MX+C
#M=80.26216284,X=given temp by us like [33],C=-1194.8953869422733
Y=80.26216284*33+(-1194.8953869422733)
Y
```

```python
x=df[['Temperature']]
y=df[['Sales']]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

```

```python
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
print('cofficient (slope):',reg.coef_[0])
print('Intercept:',reg.intercept_)
print('MSE:',mean_squared_error(y_test,y_pred))
print('R2^score:',r2_score(y_test,y_pred))
```
